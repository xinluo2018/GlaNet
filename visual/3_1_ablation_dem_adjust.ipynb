```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path_metrics_unet = 'model/trained/deeplabv3plus_mb2/deeplabv3plus_mb2_metrics_2.csv'
path_metrics_unet_dem_adjust = 'model/trained/deeplabv3plus_mb2/deeplabv3plus_mb2_dem_adjust_metrics_2.csv' 

```

```python
metrics_unet_df = pd.read_csv(path_metrics_unet)
metrics_unet_dem_adjust_df = pd.read_csv(path_metrics_unet_dem_adjust)
metrics_unet_df.tail()
metrics_unet_dem_adjust_df.tail()
```

```python
metrics_unet_df = [metrics_unet_df, metrics_unet_dem_adjust_df]
colors = ['blue', 'orange']
labels = ['without_dem_adjust', 'with_dem_adjust']  
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for i, metric_df in enumerate(metrics_unet_df):
    sns.lineplot(data=metric_df, 
                x=metric_df.index, 
                y='val_miou', 
                label=labels[i],
                color=colors[i],
                ax=ax)
ax.set_xlabel('Epochs')
ax.set_title('MIoU using dem_adjust method') 

```
