```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path_metrics_scale_1 = 'model/trained/patch_256/unet_metrics_1.csv'
path_metrics_scale_2 = 'model/trained/patch_1024/unet_metrics_1.csv'
path_metrics_unet_scales = 'model/trained/unet_256_1024/unet_metrics_1.csv' 
```

```python
metrics_scale_1_df = pd.read_csv(path_metrics_scale_1)
metrics_scale_2_df = pd.read_csv(path_metrics_scale_2)
metrics_unet_scales_df = pd.read_csv(path_metrics_unet_scales)
metrics_scale_1_df.tail()
metrics_unet_scales_df.tail()  
```

```python
metrics_models_df = [metrics_scale_1_df, metrics_scale_2_df, metrics_unet_scales_df]
colors = ['blue', 'green', 'orange']
labels = ['scale_1', 'scale_2', 'unet_scales']   
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for i, metric_df in enumerate(metrics_models_df):
    sns.lineplot(data=metric_df, 
                x=metric_df.index, 
                y='val_miou', 
                label=labels[i],
                color=colors[i],
                ax=ax)
ax.set_xlabel('Epochs')
ax.set_title('MIoU using dem_adjust method') 

```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for i, metric_df in enumerate(metrics_models_df):
    sns.lineplot(data=metric_df, 
                x=metric_df.index, 
                y='tra_miou', 
                label=labels[i],
                color=colors[i],
                ax=ax)
ax.set_xlabel('Epochs')
ax.set_title('MIoU using dem_adjust method') 

```
