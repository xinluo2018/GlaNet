```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path_metrics_u2net = 'model/trained/u2net_512/u2net_metrics_1.csv'
path_metrics_u2net_dem_adjust = 'model/trained/u2net_dem_adj_512/u2net_dem_adj_metrics_2.csv'  
```

```python
metrics_unet_df = pd.read_csv(path_metrics_u2net)
metrics_unet_dem_adjust_df = pd.read_csv(path_metrics_u2net_dem_adjust)
metrics_unet_df.tail()
# metrics_unet_dem_adjust_df.tail()
```

```python
metrics_unet_df = [metrics_unet_df, metrics_unet_dem_adjust_df]
colors = ['blue', 'orange']
labels = ['without_dem_adjust', 'with_dem_adjust']   
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for i, metric_df in enumerate(metrics_unet_df):
    sns.lineplot(data=metric_df, 
                x=metric_df.index, 
                y='val_miou', 
                label=labels[i],
                color=colors[i],
                ax=ax)
ax.set_xlabel('Epochs')
ax.set_title('MIoU using dem_adjust method') 
ax.set_ylim(0.8, 0.95)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for i, metric_df in enumerate(metrics_unet_df):
    sns.lineplot(data=metric_df, 
                x=metric_df.index, 
                y='tra_loss', 
                label=labels[i],
                color=colors[i],
                ax=ax)
ax.set_xlabel('Epochs')
ax.set_title('OA using dem_adjust method') 
# ax.set_ylim(0.6, 1)
ax.set_ylim(0., 0.1)
```
