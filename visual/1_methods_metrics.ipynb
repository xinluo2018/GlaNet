```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path_metrics_unet = 'model/trained/patch_512/unet_metrics.csv'
path_metrics_deeplabv3plus = 'model/trained/patch_512/deeplabv3plus_metrics.csv'
path_metrics_deeplabv3plus_mb2 = 'model/trained/deeplabv3plus_mb2_512/deeplabv3plus_mb2_metrics.csv'
metrics_unet_df = pd.read_csv(path_metrics_unet)
metrics_deeplabv3plus_df = pd.read_csv(path_metrics_deeplabv3plus)
metrics_deeplabv3plus_mb2_df = pd.read_csv(path_metrics_deeplabv3plus_mb2)
metrics_unet_df.tail() 
# metrics_deeplabv3plus_mb2_df.tail()
```

```python
metrics_df = [metrics_unet_df, 
              metrics_deeplabv3plus_df, 
              metrics_deeplabv3plus_mb2_df]
colors = ['blue', 'orange', 'green']
labels = ['unet', 'deeplabv3plus', 'deeplabv3plus_mb2']  
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for i, metric_df in enumerate(metrics_df):
    sns.lineplot(data=metric_df, 
                x=metric_df.index, 
                y='val_miou', 
                label=labels[i],
                color=colors[i],
                ax=ax)
ax.set_xlabel('Epochs')
ax.set_title('MIoU using patch_512') 
```
