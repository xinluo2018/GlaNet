```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path_metrics_unet = 'model/trained/patch_512/unet_metrics.csv'
path_metrics_unet_att = 'model/trained/patch_512/unet_att_metrics.csv'
```

```python
metrics_unet_df = pd.read_csv(path_metrics_unet)
metrics_unet_att_df = pd.read_csv(path_metrics_unet_att)
metrics_unet_df.tail()
metrics_unet_att_df.tail()  
```

```python
metrics_models_df = [metrics_unet_df, metrics_unet_att_df]
colors = ['blue', 'orange']
labels = ['unet', 'unet_att']  
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for i, metric_df in enumerate(metrics_models_df):
    sns.lineplot(data=metric_df, 
                x=metric_df.index, 
                y='val_miou', 
                label=labels[i],
                color=colors[i],
                ax=ax)
ax.set_xlabel('Epochs')
ax.set_title('validation MIoU using unet method') 
ax.set_ylim(0.4, 0.91) 
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for i, metric_df in enumerate(metrics_models_df):
    sns.lineplot(data=metric_df, 
                x=metric_df.index, 
                y='tra_miou', 
                label=labels[i],
                color=colors[i],
                ax=ax)
ax.set_xlabel('Epochs')
ax.set_title('training miou using unet method') 

```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for i, metric_df in enumerate(metrics_models_df):
    sns.lineplot(data=metric_df, 
                x=metric_df.index, 
                y='val_loss', 
                label=labels[i],
                color=colors[i],
                ax=ax)
ax.set_xlabel('Epochs')
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for i, metric_df in enumerate(metrics_models_df):
    sns.lineplot(data=metric_df, 
                x=metric_df.index, 
                y='tra_oa', 
                label=labels[i],
                color=colors[i],
                ax=ax)
ax.set_xlabel('Epochs')
```
